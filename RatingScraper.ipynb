```python
import pandas as pd
import numpy as np

NameCode_DF = pandas.read_csv("nameCode.csv")
FullInfoList_DF = pandas.read_csv("SummarizedCourseInfoList.csv")
```

```python
NameCode_List= NameCode_DF.values
rows = NameCode_List.shape[0]



for x in range(0,rows):
    nameToChange = NameCode_List[x,0]
    FirstName = nameToChange.split(",")[1]
    LastName = nameToChange .split(",")[0]
    newName = FirstName + " " + LastName
    newName = newName.strip()
    newName = newName.upper()
    NameCode_List[x,0] = newName
    
NameCode_List
```

```python
namesNeeded_List = FullInfoList_DF["Instructor"].tolist()
namesNeeded_List
namesNeededNoDuplicates_List = []

for name in namesNeeded_List:
    if name not in namesNeededNoDuplicates_List:
        namesNeededNoDuplicates_List.append(name)
        
namesNeededNoDuplicates_List
```

```python
teacher_codeList = []
teacher_nameList = []
for name in namesNeededNoDuplicates_List:
    
    for x in range(0,rows):
        if(NameCode_List[x,0]) == name:
            teacher_codeList.append(NameCode_List[x,1])
            teacher_nameList.append(NameCode_List[x,0])
        teacher_nameList
        
```

```python
teacherNameCode_DF = pd.DataFrame()
teacherNameCode_DF["Name"] = teacher_nameList
teacherNameCode_DF["Code"] = teacher_codeList

teacherNameCode_DF.to_csv("requiredTeacherCodes.csv", index = False)
```
